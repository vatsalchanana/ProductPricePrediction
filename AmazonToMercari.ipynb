```python
import pandas as pd
import gzip
import glob

def parse(path):
    for f in glob.glob(path+"*.json"):
        g = open(f, 'rb')
        for l in g:
            yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    if 'description' not in d.keys():
        continue
    df[i] = d
    i += 1
#     if i>100:
#         break
  return pd.DataFrame.from_dict(df, orient='index')
```

```python
df = getDF('./')
print(df.info())
print(df.describe())

# df.describe()
# df
```

```python
df=df.drop(columns=['related', 'imUrl'])
df2=df.dropna(subset=['price'])
df2.to_csv('Descriptions_metadata.csv', encoding='utf-8', index=False)
```
